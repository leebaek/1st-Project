# 월별 승하차 csv파일 병합(merge)

```python
import pandas as pd
import glob
import os
import sys
```

## 경로에 있는 csv파일을 새로운 csv파일로 합침

```python
input_path = sys.argv[1]
output_file = sys.argv[2]

all_files = glob.glob(os.path.join(input_path,'C:/Python/project01/월별 지하철 승하차 15-19/CARD_*.csv'))

all_data_frames = []

for file in all_files:
    data_frame = pd.read_csv(file, index_col=None, encoding = 'utf-8')
    all_data_frames.append(data_frame)

data_frame_concat = pd.concat(all_data_frames, axis=0, ignore_index=True)
data_frame_concat.to_csv("subway_merge", index = False)
```
